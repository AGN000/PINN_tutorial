```python
import numpy as np
import matplotlib.pyplot as plt

# Define exact function
def f_exact(x):
    return x**2

# Define Fourier cosine series approximation
def f_fourier_cosine(x, N):
    result = 1/3  # a0/2 term
    for n in range(1, N+1):
        an = (4 * (-1)**n) / (n**2 * np.pi**2)
        result += an * np.cos(n * np.pi * x)
    return result

# Domain
x = np.linspace(0, 1, 1000)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x, f_exact(x), 'k-', label='Exact: $x^2$', linewidth=2)

# Compare for different N
for N in [1, 5, 10, 50]:
    y_approx = np.array([f_fourier_cosine(xi, N) for xi in x])
    plt.plot(x, y_approx, label=f'N={N}')

plt.title('Fourier Cosine Series Approximation of $x^2$ on [0, 1]')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
